1. Genere una onda sinusoidal de duraci칩n 1s., amplitud 洧냢 = 1 y frecuencia de muestreo 洧녭 = 1000洧냩洧녾. Var칤e la frecuencia fundamental del tono 洧녭 = [100, 200, 300]洧냩洧녾 y reproduzca los comandos usando una primitiva adecuada en Julia.

```julia
using Plots
using WAV
using Sound
using Plots

amplitud = 1.0      # Amplitud de la onda
fase = 0.0          # Fase inicial de la onda
duracion = 1.0      # Duraci칩n de la se침al en segundos
fs = 1000           # N칰mero de puntos de fs por segundo

# Crear vector de tiempo
t = 0:1/fs:duracion

# Frecuencias a variar
frecs = [100, 200, 300]

# Crear un plot con subplots
p = plot(layout=(length(frecs), 1))

# Loop a trav칠s de cada frecuencia y graficar la onda correspondiente
for (i, f0) in enumerate(frecs)
    y_loop = amplitud * sin.(2 * f0 * t)
    plot!(p, t, y_loop, label="FREC - $f0 HZ", subplot=i)
end

# Mostrar el gr치fico final
display(p)
```


2. Represente la onda sinusoidal del punto 1 para f0 = [5, 10] Hz, compruebe gr치ficamente el valor del periodo (T0). Represente tambi칠n el m칩dulo de la FFT y compruebe el valor de f0.


```julia
fs = 150           # N칰mero de puntos de fs por segundo

# Crear vector de tiempo
t = 0:1/fs:duracion

# Frecuencias a variar
frecs = [5, 10]

# Crear un plot con subplots
pp = plot(layout=(length(frecs), 1))

signals = []

# Loop a trav칠s de cada frecuencia y graficar la onda correspondiente
for (i, f0) in enumerate(frecs)
    y_loop = amplitud * sin.(2 * f0 * t)
    push!(signals, y_loop)
    period = (1/f0)
    plot!(pp, t, y_loop, label="Frecuency $f0 HZ", subplot=i)
    scatter!(pp, [period:period:maximum(t)], [0.0], markercolor=:red, markersize=5, label="Period $(1/f0) Hz", subplot=i)
end

# Mostrar el gr치fico final
display(pp)
```

Hemos marcado en puntos rojos, el periodo de la se침al, repetido T, 2T, 3T... hasta m치ximo t.
Ahora procedemos a representar el m칩dulo de la FFT

```julia
using FFTW
plotlyjs()

p2 = plot(layout=(length(signals), 1))

for (i, signal) in enumerate(signals)

  y_fft = fftshift(fft(signal))
  y_fft_modulo = abs.(y_fft)
  n = length(signal)
  frequencies = fftshift(fftfreq(n))
  print(frequencies)
  plot!(p2, frequencies, y_fft_modulo, xlabel="Frequency", ylabel="|FFT|", label="FFT Module", legend=:topright, subplot=i, hover=frequencies)
end

display(p2)
```

```julia
# TEST

using MAT
using Plots
using FFTW

# Extraer la se침al y la frecuencia de muestreo
signal = datos_matlab["trama3"]
fs = 11000

# Calcular la FFT de la se침al
N = length(signal)
frequencies = fftfreq(N, fs)
fft_result = fft(signal)

# Calcular el m칩dulo de la FFT
fft_modulo = abs.(fft_result)

# Encontrar la frecuencia fundamental (f0)
f0_index = argmax(fft_modulo[2:end]) + 1
f0 = frequencies[f0_index]
print(f0)

# Encontrar las frecuencias formantes (por ejemplo, las primeras tres)
num_formants = 10
formants_indices = sortperm(fft_modulo[2:end], rev=true)[1:num_formants]
formants_frequencies = frequencies[formants_indices .+ 1]

p3 = plot(layout=(1, 1))

# Crear el gr치fico
plot!(p3, frequencies, fft_modulo, xlabel="Frequency (Hz)", ylabel="|FFT|", label="FFT Modulus")
scatter!([f0], [maximum(fft_modulo)], color=:red, label="Fundamental Frequency (f0)")
scatter!(frequencies[formants_indices .+ 1], fft_modulo[formants_indices .+ 1], color=:green, label="Formant Frequencies")

# Muestra el gr치fico
display(p3)

sound(signal, fs)
```

c. Calcule el espectro de frecuencia de dichas ventanas, usando la FFT y compare los m칩dulos de los espectros de ambas ventanas represent치ndolos lado a lado.
   

```julia
using Plots
using LinearAlgebra
include("functions.jl")
# comparamos diferentes ventanas, como por ejemplo, con una rectangular

Rect_Frec = fft(rectangular_window)
Hamm_Frec = fft(h_window)  # Asumo que h_window es la ventana de Hamming

ejex = LinRange(0, 1, length(Rect_Frec))

# Graficar
plot1 = plot(ejex[1:div(end, 2)], 20 * log.(abs.(Rect_Frec[1:div(end, 2)]) / maximum(abs.(Rect_Frec[1:div(end, 2)]))),
    title="Rectangular", xlabel="Hz/fs", ylabel="Amplitud(dB)", legend=false)

plot2 = plot(ejex[1:div(end, 2)], 20 * log.(abs.(Hamm_Frec[1:div(end, 2)]) / maximum(abs.(Hamm_Frec[1:div(end, 2)]))),
    title="Hamming", xlabel="Hz/fs", ylabel="Amplitud(dB)", legend=false)

plot(plot1, plot2, layout=(2, 1))
```
